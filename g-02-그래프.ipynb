## 그래프 그리기


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
n = 10
y = np.random.randn(n)
plt.plot(range(n), y)
plt.xlabel('시간')
plt.ylabel('랜덤 값')
```

```python
x = np.random.randint(1,7, 10000)
plt.hist(x, bins = 6, width=0.6)
plt.show()
```

```python
z = 0
n = 4 # number of indices thrown
for i in range(n):
    x = np.random.randint(1,7, 100000)
    z = z + x
plt.hist(z,bins = n*5+1, width=0.6)
plt.show()
```

```python
x = np.random.randn(100000)
plt.hist(x, bins = 100, width=0.03)
plt.show()
```

```python
n = 10000
x = np.random.rand(n)
y = np.random.rand(n)
plt.figure(figsize=(6,6))
plt.scatter(x,y, s=1)
xx=np.linspace(0,1,100)
plt.plot(xx, (1-xx*xx)**0.5, c='r')

pi = ((x**2 + y**2) < 1).mean()*4  
pi
```
