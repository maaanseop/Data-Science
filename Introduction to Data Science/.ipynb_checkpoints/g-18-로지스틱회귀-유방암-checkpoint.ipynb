```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
%matplotlib inline
```

```python
#임의로 선정한 시간 합격 공부 시간, 불합격 공부 시간
pass_time = [8,9, 9, 9.5, 10, 12, 14, 14.5, 15, 16, 16, 16.5, 17, 17, 17, 17.5,20, 20, 20]
fail_time = [1, 2, 2.1,2.6,2.7,2.8,2.9, 3,3.2,3.4,3.5,3.6,3, 5,5.2,5.4]
```

```python
#두 데이터를 옆으로 묶는 것
X = np.hstack((pass_time,fail_time))
X
```

```python
#합격자 수만큼 1을 배정하고 불합격자 수만큼 0을 배정
y1=[1]*len(pass_time)
y0=[0]*len(fail_time)
y = np.hstack((y1,y0))
y
```

```python
#그림 그린 결과 확인
fig = plt.figure(figsize=(8,6))
plt.xlim(0, 21)
plt.ylim(-0.1, 1.1)
plt.xlabel("Study time")
plt.ylabel("Pass rate")
plt.scatter(X, y)
plt.show()
```

```python
#생성자 생성
model = LogisticRegression()
#모델 핏팅, 1차원을 2차원으로 바꿈
model.fit(X.reshape(-1, 1),y)

print(model.coef_)
print(model.intercept_)
```

```python
model.predict([[4.8]])
```

```python
model.predict_proba([[13]])
```

```python
def logreg(z):
    return 1 /(1+np.exp(-z))

fig = plt.figure(figsize=(8,6))
plt.xlim(0, 20)
plt.ylim(-0.1, 1.1)
plt.xlabel('Study time')
plt.ylabel('Pass rate')
plt.scatter(X, y, s=50)

XX = np.linspace(0.5,21, 100)
yy = logreg(model.coef_*XX + model.intercept_)[0]
plt.plot(XX, yy, c='r')
# plt.show()
```

```python
def logreg(z):
    return 1 /(1+np.exp(-z))

fig = plt.figure(figsize=(8,6))
plt.xlim(-10, 10)
plt.ylim(-0.1, 1.1)
a=2 #a라는 변수는 기울기라고 생각 a가 작으면 원만하게 바뀌고 커지면 스무스하게 바뀐다.
b=-5 #b는 오른쪽 왼쪽으로 움직이는 역할 Ex)5를 주면 -2.5까지 내려감
XX = np.linspace(-10, 10, 100)
yy = logreg(a*XX + b)
plt.plot(XX, yy, c='r')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold   # K-fold cross validation
from sklearn import metrics
```

```python
data = pd.read_csv('data/breast_cancer.csv')
print(data.shape)
```

```python
data.head()
```

```python
data['diagnosis'] = data['diagnosis'].map({'M':1,'B':0})#기존의 값을 매핑해서 바꿔주는 함수 map. ex)M은 1로 B는 0으로
df = data[["diagnosis", "radius_mean", "texture_mean"]]
df.head()
```

```python
model=LogisticRegression()
y = df["diagnosis"]
```

```python
features = ["radius_mean"]#반경
X = df[features]
np.random.seed(11)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

model.fit(X_train, y_train)
print("Score: {0:.2%}".format(model.score(X_test, y_test)))
```

```python
X_test.shape
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
plt.figure(figsize=(8,4))
plt.scatter(X_train, y_train, s=1)
plt.xlabel("radius_mean")
plt.ylabel("M(1) or B(0)")

def logreg(x):
    return 1 / (1 + np.exp(-x))

XX = np.linspace(7.5, 40, 100)
plt.plot(XX, logreg(model.coef_ * XX + model.intercept_)[0], c='r')
```

```python
features = ["texture_mean"]#쓸만한 변수가 아니다. 50프로가 기본
X = df[features]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

model.fit(X_train, y_train)
print("Score: {0:.2%}".format(model.score(X_test, y_test)))
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
plt.figure(figsize=(8,4))
plt.scatter(X_train, y_train, s=1)
plt.xlabel("texture_mean")
plt.ylabel("M(1) or B(0)")

XX = np.linspace(7.5, 40, 100)
plt.plot(XX, logreg(model.coef_ * XX + model.intercept_)[0], c='r')
```

```python
features = ["radius_mean", "texture_mean"]

X = df[features]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

model.fit(X_train, y_train)
print("Score: {0:.2%}".format(model.score(X_test, y_test)))
```

```python
markers = ['o', 'v']
label = ["B", "M"]
for i in range(2):
    xs = X_train["radius_mean"][y_train == i]
    ys = X_train["texture_mean"][y_train == i]
    plt.scatter(xs, ys, marker=markers[i], s=10)
plt.legend(label)
plt.xlabel("radius_mean")
plt.ylabel("texture_mean")
```

```python
# 모든 속성을 다 사용하여 로지스틱 회귀 분석을 수행
# 필요없는 값은 다 빼버림, 결과값도 뺌
data.drop('id',axis=0,inplace=True)
data.drop('Unnamed: 32',axis=1,inplace=True)
X=data.drop('diagnosis', axis=1)
```

```python
np.random.seed(11)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
model_all = LogisticRegression()
model_all.fit(X_train, y_train)
print("Score: {0:.2%}".format(model_all.score(X_test, y_test)))
```

```python
# 모든 속성을 다 사용한 경우의 kNN의 성능
from sklearn.neighbors import KNeighborsClassifier  
for i in range(1,21,2):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    print("K = ", i, "-> Score: {0:.2%}".format(knn.score(X_test, y_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
print("결정트리 -> {:.2%}".format(tree.score(X_test, y_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=1000)
rfc.fit(X_train, y_train)
print("랜덤포레스트 : {:.2%}".format(rfc.score(X_test, y_test)))
```
